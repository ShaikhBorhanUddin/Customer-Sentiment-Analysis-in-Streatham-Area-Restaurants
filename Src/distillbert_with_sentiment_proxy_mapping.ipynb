```python
import pandas as pd
from transformers import pipeline
import torch

# Load your dataset
combined_df = pd.read_csv('combined_df.csv')

# Step 1: Load DistilBERT sentiment pipeline
device = 0 if torch.cuda.is_available() else -1
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english", device=device)

# Step 2: Apply DistilBERT on non-null review texts
def classify_distilbert_sentiment(text):
    result = classifier(text[:512])[0]  # truncate to 512 tokens
    return 'positive' if result['label'] == 'POSITIVE' else 'negative'

combined_df['bert_sentiment'] = combined_df['text'].dropna().apply(classify_distilbert_sentiment)

# Step 3: Proxy sentiment mapping for null text
def proxy_sentiment_from_stars(star):
    if star >= 4:
        return 'positive'
    elif star == 3:
        return 'neutral'
    elif pd.notnull(star):
        return 'negative'
    else:
        return None

combined_df['proxy_sentiment'] = combined_df.apply(
    lambda row: proxy_sentiment_from_stars(row['stars']) if pd.isnull(row['text']) else None,
    axis=1
)

# Step 4: Merge BERT and proxy sentiments
combined_df['final_sentiment'] = combined_df['bert_sentiment'].combine_first(combined_df['proxy_sentiment'])

# Step 5: Aggregate sentiment by restaurant using 'title' column
sentiment_summary = combined_df.groupby('title')['final_sentiment'].value_counts().unstack(fill_value=0)  # Changed 'restaurantName' to 'title'
sentiment_summary['total'] = sentiment_summary.sum(axis=1)
sentiment_summary['positive_ratio'] = sentiment_summary['positive'] / sentiment_summary['total']

# Sort by positive ratio
sentiment_summary_sorted = sentiment_summary.sort_values(by='positive_ratio', ascending=False)

# Show top results
print(sentiment_summary_sorted)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming sentiment_summary_sorted is already calculated

# Select top 12 restaurants
top_12_restaurants = sentiment_summary_sorted.head(12)

# Create a 4x3 subplot grid
fig, axes = plt.subplots(4, 3, figsize=(15, 15))  # Adjust figsize as needed

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Iterate through each restaurant and create a pie chart in a subplot
for i, restaurant in enumerate(top_12_restaurants.index):
    # Get sentiment counts for the current restaurant
    sentiments = top_12_restaurants.loc[restaurant, ['positive', 'negative']]

    # Create pie chart in the current subplot
    axes[i].pie(sentiments, labels=['Positive', 'Negative'], autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f'Sentiment for {restaurant}')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import textwrap  # Import textwrap for wrapping text

# Select top 12 restaurants
top_12_restaurants = sentiment_summary_sorted.head(12)

# Extract restaurant names and positive ratios
restaurants = top_12_restaurants.index
positive_ratios = top_12_restaurants['positive_ratio']

# Generate distinct colors for each bar
colors = plt.cm.get_cmap('viridis', len(restaurants))  # You can choose other colormaps

# Create bar chart
plt.figure(figsize=(14, 6))  # Adjust figure size as needed
bars = plt.bar(restaurants, positive_ratios, color=colors(np.arange(len(restaurants))))

# Display values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.001, round(yval, 2), ha='center', va='bottom')

# Customize chart appearance
plt.xlabel('Restaurant')
plt.ylabel('Positive Ratio')
plt.title('Positive Sentiment Ratios for Top 12 Restaurants')

# Wrap x-axis labels for better readability
labels = [textwrap.fill(str(label), 15) for label in restaurants]  # Wrap labels with textwrap
plt.xticks(restaurants, labels, rotation=0, ha='center', fontsize=8)  # Set wrapped labels

plt.tight_layout()

# Show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import textwrap  # Import textwrap for wrapping text

# Assuming sentiment_summary_sorted and df are already calculated/loaded

# Select top 12 restaurants
top_12_restaurants_names = sentiment_summary_sorted.head(12).index

# Filter the original dataframe to include only the top 12 restaurants
top_12_restaurants_df = df[df['title'].isin(top_12_restaurants_names)]

# Aggregate categories for top 12 restaurants
all_categories = []
for i in range(10):
    category_column = f'categories/{i}'
    categories = top_12_restaurants_df[category_column].dropna().tolist()
    all_categories.extend(categories)

# Count category occurrences
category_counts = pd.Series(all_categories).value_counts()

# Select top categories (e.g., top 10)
top_categories = category_counts.head(10)

# Create bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(top_categories.index, top_categories.values)  # Store the bar objects
plt.xlabel('Category')
plt.ylabel('Number of Occurrences')
plt.title('Frequently mentioned Tags in reviews')

# Wrap x-axis labels for better readability
labels = [textwrap.fill(label, 15) for label in top_categories.index]  # Adjust wrap width (15) as needed
plt.xticks(top_categories.index, labels, rotation=0, ha='center', fontsize=8)  # Set wrapped labels

plt.tight_layout()

# Add count values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, int(yval), ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from transformers import pipeline
import torch

# Load DistilBERT sentiment pipeline
device = 0 if torch.cuda.is_available() else -1
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english", device=device)

# Choose the individual reviews with their row indices
sample_texts_with_indices = [
    (df['text'].dropna().iloc[2301], 2301),  # First review, index 0
    (df['text'].dropna().iloc[4022], 4022),  # Second review, index 3022
    (df['text'].dropna().iloc[999], 999)  # Third review, index 4612
]

# Create a 1x3 subplot
fig, axes = plt.subplots(1, 3, figsize=(18, 4))  # Adjust figsize as needed

# Iterate through the reviews and create visualizations
for i, (sample_text, row_index) in enumerate(sample_texts_with_indices):
    result = classifier(sample_text[:512])[0]  # Truncate to 512 tokens
    labels = ['Negative', 'Positive']
    values = [1 - result['score'], result['score']] if result['label'] == 'POSITIVE' else [result['score'], 1 - result['score']]

    # Create bar chart in the current subplot
    bars = axes[i].bar(labels, values, color=['red', 'green'])
    axes[i].set_title(f"DistilBERT Sentiment Breakdown (Entry Row: {row_index})\nLabel: {result['label']}, Score: {result['score']:.2f}")
    axes[i].set_ylabel('Score')
    axes[i].set_ylim(0, 1.2)

    # Add values on top of bars
    for bar in bars:
        yval = bar.get_height()
        axes[i].text(bar.get_x() + bar.get_width() / 2, yval + .005,
                     '{:.2f}'.format(yval), ha='center', va='bottom')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
