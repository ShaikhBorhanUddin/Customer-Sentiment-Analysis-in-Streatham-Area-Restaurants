```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io
```

```python
# Combine all uploaded CSVs into a single DataFrame
dataframes = []

for filename in uploaded.keys():
    df = pd.read_csv(io.StringIO(uploaded[filename].decode('utf-8')))
    dataframes.append(df)

# Combine into one DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
# Assuming your DataFrame is named 'combined_df' and has columns 'title' and 'stars'
five_star_counts = combined_df.groupby(['title', 'stars'])['stars'].size().reset_index(name='count')
five_star_counts = five_star_counts[five_star_counts['stars'] == 5]

# Display the results
print(five_star_counts)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import textwrap

plt.figure(figsize=(14, 8))
plt.bar(five_star_counts['title'], five_star_counts['count'])
plt.xlabel('Title')
plt.ylabel('Number of 5-Star Ratings')
plt.title('Distribution of 5-Star Ratings by Title')

# Wrap x-axis labels into two lines
labels = [textwrap.fill(label, 15) for label in five_star_counts['title']] # Adjust 15 to change wrap length
plt.xticks(range(len(labels)), labels, rotation=0, ha='center')

# Display bar counts on top
for i, v in enumerate(five_star_counts['count']):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Group data by title and stars, then get counts
star_counts_by_title = combined_df.groupby(['title', 'stars'])['stars'].size().unstack(fill_value=0)

# Create a figure and subplots
fig, axes = plt.subplots(3, 4, figsize=(16, 12))  # 3 rows, 4 columns

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Iterate through titles and create pie charts
for i, title in enumerate(star_counts_by_title.index):
    # Get star counts for the current title
    counts = star_counts_by_title.loc[title]

    # Create pie chart in the current subplot
    axes[i].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f'Star Ratings for {title}')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import folium
import pandas as pd

# Create a map centered on the average latitude and longitude
map_center = [combined_df['location/lat'].mean(), combined_df['location/lng'].mean()]
my_map = folium.Map(location=map_center, zoom_start=12)  # Adjust zoom_start as needed

# Add markers for each location with 5-star ratings
for index, row in five_star_counts.iterrows():
    folium.Marker(
        location=[combined_df.loc[combined_df['title'] == row['title'], 'location/lat'].iloc[0],
                  combined_df.loc[combined_df['title'] == row['title'], 'location/lng'].iloc[0]],
        popup=f"{row['title']} ({row['count']} 5-star ratings)",
        icon=folium.Icon(color='green')  # Customize marker color if desired
    ).add_to(my_map)

# Display the map
my_map
```
